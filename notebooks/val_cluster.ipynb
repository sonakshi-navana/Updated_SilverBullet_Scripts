```python
import pickle

with open(pickle_folder_path+'val_dict.pickle', 'rb') as file:
    dict = pickle.load(file)
```

```python
import pandas as pd 
```

```python
import numpy as np 
```

```python
df= pd.DataFrame()
df["ID"]=dict.keys()
df["emb"]=dict.values()
df.head()
```

```python
df['emb_dim'] = df['emb'].apply(lambda x: x.shape if type(x)==np.ndarray else None)
df.head()
```

```python
df['spk_id'] = df['ID'].str.split('_').str[0]
df.head()
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
df['emb'] = df['emb'].apply(lambda x: x.flatten())
```

```python
df.head()
```

```python
df['emb_dim'] = df['emb'].apply(lambda x: x.shape if type(x)==np.ndarray else None)
df.head()
```

get centroids

```python

unique_ids = df['spk_id'].unique()
print(unique_ids)

centroids = {}

# iter over each id 
for id_val in unique_ids:
    # get emb values for the id
    emb_values = df[df['spk_id'] == id_val]['emb'].tolist()
    # fit the k-means 
    kmeans = KMeans(n_clusters=1, random_state=0).fit(emb_values)
    # centroid calculation
    centroids[id_val] = kmeans.cluster_centers_


centroids
```

```python
centroid_df= pd.DataFrame()
centroid_df["spk_id"] = centroids.keys()
centroid_df["centroid"]= centroids.values()
centroid_df.head()
```

```python
centroid_df['centroid'] = centroid_df['centroid'].apply(lambda x: x.flatten())
centroid_df['cen_dim'] = centroid_df['centroid'].apply(lambda x: x.shape if type(x)==np.ndarray else None)
centroid_df.head()
```

```python
centroid_df.to_csv(output_folder+"centroids.csv", index=False)
```

```python
from scipy.spatial import distance
```

unval distances- rank ordered

```python
# Create an empty list to store the distances
distances = []


for index, row in unval_df.iterrows():
    
    spk_id = row['spk_id']
    emb = row['emb']
    
    # Get the corresponding centroid for the current spk_id
    cent = centroid_df.loc[centroid_df['spk_id'] == spk_id]['centroid'].values[0]
    
    
    dist = distance.cosine(emb, cent)
    
    
    distances.append(dist)

unval_df['distance'] = distances

unval_df.head()
```

```python
unval_df = unval_df.sort_values(by=['spk_id', 'distance'])
unval_df.reset_index(drop=True)
unval_df.head()
```

MODULAR CODE

```python
for spk_id in unique_spk_ids:
    emb1 = df_first.loc[df_first['spk_id'] == spk_id, 'emb'].values[0]

    # Initialize an empty list to store cosine similarities
    cosine_similarities = []

    # Iterate through alltables
    for index, row in alltables[spk_id].iterrows():
        emb2 = row['emb']
        #cosine_similarity = np.dot(emb1, emb2) / (np.linalg.norm(emb1) * np.linalg.norm(emb2))
        cos_sim = cosine_similarity(emb1.reshape(1, -1), emb2.reshape(1,-1))[0][0]
        cosine_similarities.append(cos_sim)
    alltables[spk_id]= alltables[spk_id].assign(similarity= cosine_similarities)
    alltables[spk_id]=alltables[spk_id].drop(columns=["emb_dim","spk_id"])
    alltables[spk_id].to_csv(output_folder+spk_id+".csv", index=False)
```
